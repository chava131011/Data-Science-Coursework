# Análisis exploratorio de datos (EDA)

El análisis exploratorio de datos o Exploratory Data Analysis(EDA) es un enfoque para analizar conjuntos de datos para resumir sus principales características. Se utiliza para comprender los datos, obtener un contexto al respecto, comprender las variables y las relaciones entre ellas y formular hipótesis que podrían ser útiles a la hora de construir modelos predictivos.

Al visualizar de forma eficaz las variables de un conjunto de datos y sus relaciones,  un científico de datos puede comprender rápidamente tendencias, valores atípicos y patrones. Esta comprensión se puede utilizar para contar una historia, impulsar decisiones y crear modelos predictivos.

**En esta lección, aprenderemos cómo realizar EDA utilizando la visualización de datos. Específicamente, nos centraremos en funciones asociadas a Pandas para graficar.**

<div class="alert alert-success">
Mostrar la matriz de correlación

Colormaps:
https://matplotlib.org/stable/users/explain/colors/colormaps.html

</div>



```python
corr = df.corr(numeric_only=True)
corr
```

```python
import matplotlib.pyplot as plt
plt.matshow(corr, cmap='RdYlGn', vmin=-0.5, vmax=1)
plt.title('Matriz de correlación', pad=20)
plt.colorbar()
plt.xticks(range(len(corr.columns)), list(corr.columns), rotation=90)
plt.yticks(range(len(corr.columns)), list(corr.columns))

for (i, j), val in np.ndenumerate(corr.values):
    plt.text(j, i, f'{val:.2f}', ha='center', va='center', color='white', fontsize=9)

plt.grid(False)

plt.show()
```
